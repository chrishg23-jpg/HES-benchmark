```python
import numpy as np

# Parameters
num_qubits = 5
num_steps = 30  # Extended stage
amplitudes = np.array([0.99, 1.01, 1.004, 1.004, 1.002])
mutation_rate = 0.002
law_band = (0.997, 1.007)

# Storage
codex = []
declared = False

# Simulation loop
for t in range(num_steps):
    frame_codex = []

    for q in range(num_qubits):
        if not declared:
            amplitudes[q] += np.random.uniform(-mutation_rate, mutation_rate)
        frame_codex.append(f"Codex [t={t}]: Qubit_{q} amplitude={round(amplitudes[q],6)}")

    entanglement_variance = round(np.std(amplitudes), 6)
    frame_codex.append(f"Codex [t={t}]: Entanglement variance = {entanglement_variance}")

    # Declaration check
    if not declared and all(law_band[0] <= a <= law_band[1] for a in amplitudes) and entanglement_variance < 0.005:
        declared = True
        frame_codex.append("Codex [declaration]: Quantum Emergent Law stabilized â€” eigenstate recognized")
        frame_codex.append("Codex [declaration]: HES_SystemsComplete")

    codex.extend(frame_codex)

# Output
for entry in codex:
    print(entry)
```
